## Building the Vectorization

```python
import pandas as pd
import numpy as np
from tensorflow.keras.layers import TextVectorization

df = pd.read_csv('train.csv')
```

df=df.iloc[:70000,:]


# Initializing the vectorization part
x = df["comment_text"]
y=df.iloc[:,2:]

max_features = 50000
vectorizer = TextVectorization(max_tokens=max_features,
                               output_sequence_length=1000,
                               output_mode='int')
vectorizer.adapt(x.values)
vectorized_text = vectorizer(x.values)

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(50000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(6)

batchx,batchy=dataset.as_numpy_iterator().next()




train = dataset.take(int(len(dataset)*0.7))
val = dataset.skip(int(len(dataset)*0.7)).take(int(len(dataset)*0.2))
test = dataset.skip(int(len(dataset)*0.9)).take(int(len(dataset)*0.1))



train_generator = train.as_numpy_iterator()
```

## Evaluating the model using differnt metrics

```python
from tensorflow.keras.metrics import Precision,Recall,CategoricalAccuracy
```

```python
pre=Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    xdata,ydata = batch
    yhat = model.predict(xdata)
    ydata = ydata.flatten()
    yhat = yhat.flatten()

    pre.update_state(ydata,yhat)
    re.update_state(ydata,yhat)
    acc.update_state(ydata,yhat)
```

```python
print("Precision",pre.result().numpy())
print("Recall",re.result().numpy())
print("accuracy",acc.result().numpy())
```

```python
def score_comment(comment):
    vectorized_comment = vectorizer(comment)
    result = model.predict(np.expand_dims(vectorized_comment,0))

    text = ''
    for idx,col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col,result[0][idx]>0.5)
    return text
```

```python
interface = gr.Interface(
    fn=score_comment,
    inputs=gr.Textbox(lines=2, placeholder='Comment To Score'),
    outputs=gr.Text()
)
```

```python
interface.launch(share=True)
```
